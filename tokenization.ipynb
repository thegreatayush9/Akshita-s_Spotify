# Stemming

Stemming is the process of reducing a word to its word stem that affixes to suffixes and prefixes or to the roots of words known as a lemma. Stemming is important in natural language understanding (NLU) and natural language processing (NLP).

```python
## Classification Problem
## Comments of product is a positive review or negative review
## Reviews----> eating, eat,eaten [going,gone,goes]--->go

words=["eating","eats","eaten","writing","writes","programming","programs","history","finally","finalized"]
```

```python
## Lets see how we can find out the stem of these words
```

##PorterStemmer

```python
from nltk.stem import PorterStemmer
stemming  = PorterStemmer()
for word in words:
    print(word + "----->" + stemming.stem(word))
```

## RegexpStemmer class

NLTK has RegexpStemmer class with the help of which we can easily implement Regular Expression Stemmer algorithms. It basically takes a single regular expression and removes any prefix or suffix that matches the expression. Let us see an example

```python
from nltk.stem import RegexpStemmer
stemmer = RegexpStemmer('ing$|s$|e$|able$', min=4)
for word in words:
    print(word + "----->" + stemmer.stem(word))
```

```python
stemmer1 = RegexpStemmer('ing$|s$|e$|able$', min=4)
print(stemmer1.stem('ingeating'))
```

```python
stemmer2 = RegexpStemmer('ing|s$|e$|able$', min=4)
print(stemmer2.stem('ingeating'))
```

##Snowball Stemmer

It is a stemming algorithm which is also known as the Porter2 stemming algorithm as it is a better version of the Porter Stemmer since some issues of it were fixed in this stemmer.

```python
from nltk.stem import SnowballStemmer
stemmer = SnowballStemmer('english', ignore_stopwords=False)
for word in words:
    print(word + "----->" + stemmer.stem(word))
```

```python
stemming.stem('fairly'),stemming.stem('sportingly')
```

```python
snowballStemmer = SnowballStemmer('english', ignore_stopwords=False)
snowballStemmer.stem('fairly'),snowballStemmer.stem('sportingly')
```

#Wordnet Lemmatizer

Lemmatization technique is like stemming. The output we will get after lemmatization is called ‘lemma’, which is a root word rather than root stem, the output of stemming. After lemmatization, we will be getting a valid word that means the same thing.

NLTK provides WordNetLemmatizer class which is a thin wrapper around the wordnet corpus. This class uses morphy() function to the WordNet CorpusReader class to find a lemma. Let us understand it with an example −

```python
from nltk.stem import WordNetLemmatizer
lemmatizer = WordNetLemmatizer()
```

```python
nltk.download('wordnet')
```

```python
'''
## POS-
Noun-n
verb-v
adjective-a
adverb-r
'''
lemmatizer.lemmatize("going", pos='v')
```

```python
for word in words:
    print(word + "----->" + lemmatizer.lemmatize(word, pos='v'))
```

```python
lemmatizer.lemmatize('fairly'),lemmatizer.lemmatize('sportingly')
```

## Parts of Speech Tagging

Part-of-Speech (POS) tagging is a foundational task in Natural Language Processing (NLP) that involves assigning each word in a sentence to its respective grammatical category, such as nouns, verbs, adjectives, etc. This process enables machines to understand sentence structure and the role of each word.

For example, consider this sentence:

*“Jessica loves painting beautiful landscapes.”*

After applying POS tagging using NLTK, the sentence is tagged as:

* Jessica → NNP (Proper noun,
singular)
* loves → VBZ (Verb, third person singular present)
* painting → VBG (Verb, gerund or present participle)
* beautiful → JJ (Adjective)
* landscapes → NNS (Noun, plural)

```python
# CC coordinating conjunction
# CD cardinal digit
# DT determiner
# EX existential there (like: “there is” … think of it like “there exists”)
# FW foreign word
# IN preposition/subordinating conjunction
# JJ adjective ‘big’
# JJR adjective, comparative ‘bigger’
# JJS adjective, superlative ‘biggest’
# LS list marker 1)
# MD modal could, will
# NN noun, singular ‘desk’
# NNS noun plural ‘desks’
# NNP proper noun, singular ‘Harrison’
# NNPS proper noun, plural ‘Americans’
# PDT predeterminer ‘all the kids’
# POS possessive ending parent’s
# PRP personal pronoun I, he, she
# PRP$ possessive pronoun my, his, hers
# RB adverb very, silently,
# RBR adverb, comparative better
# RBS adverb, superlative best
# RP particle give up
# TO, to go ‘to’ the store.
# UH interjection, errrrrrrrm
# VB verb, base form take
# VBD verb, past tense took
# VBG verb, gerund/present participle taking
# VBN verb, past participle taken
# VBP verb, sing. present, non-3d take
# VBZ verb, 3rd person sing. present takes
# WDT wh-determiner which
# WP wh-pronoun who, what
# WP$ possessive wh-pronoun whose
# WRB wh-abverb where, when
```

#Named Entity Recognition (NER)

Named Entity Recognition (NER) is an exciting and vital part of NLP, helping machines understand and interpret human language more meaningfully. Imagine transforming a simple sentence like this:

*“Amelia Johnson, a software engineer born in Paris, recently joined Google as a product manager.”*

By applying NER, the sentence can be broken down into its core entities:

* Person: “Amelia Johnson”
* Location: “Paris”
* Organization: “Google”

### Examples of Common Entity Types:

* Person Names: Albert Einstein
* Organizations: GeeksforGeeks
* Locations: Paris
* Dates and Times: 5th May 2025
* Quantities and Percentages: 50%, $100

```python
sentence="The Eiffel Tower was built from 1887 to 1889 by Gustave Eiffel, whose company specialized in building metal frameworks and structures."
```

```python
words = nltk.word_tokenize(sentence)
words
```

```python
tagged_elements = nltk.pos_tag(words)
```
